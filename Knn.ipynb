```python
ft=['radius_mean', 'texture_mean', 'perimeter_mean','area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
```

```python
le=LabelEncoder()
```

```python
y=le.fit_transform(df['diagnosis'])
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,y_train.shape,y_test.shape,y_test.shape
```

```python
ke=KNeighborsClassifier()
```

```python
ke.fit(x_train,y_train)
```

```python
y_pred =ke.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
confusion=confusion_matrix(y_test,y_pred)

print('confusion_matrix:\n %s'%confusion)
```

```python
classification=classification_report(y_test,y_pred)

print('classification_report:\n %s'%classification)
```

```python
sns.heatmap(confusion,annot=True,fmt='d',cmap='Blues')
plt.title('Confusion matrix')
plt.xlabel('Predicted')
plt.ylabel('actual')
plt.show()
```
